```python
x=np.arange(7)
```

```python
y=np.arange(4,11)
for i in range(7):
    y[i]=y[i]+(i*i)
```

```python
data={'x':x,'y':y}
df=pd.DataFrame.from_dict(data)
```

```python
plt.plot(x,y)
b=3
plt.axvline(x=b,linestyle='dashed', color='red', 
                label= "x value at, {}".format(b)) # to get a point that has to be highlighted
```

```python
jp=sns.JointGrid(x='x', y='y', data=df)
reg_plot=jp.plot(sns.regplot, sns.distplot)
```

```python
#pairplot
sns.pairplot(df)
```

```python
x_axis = np.arange(0, len(df.x))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)

plt.title("box_plot_in_deail")
plt.xlabel("X")
plt.ylabel("Y")

plt.xlim(-0.75, len(df.x)-.25)
plt.ylim(0, max(df.y) + 5)

plt.bar(x_axis, df.y, facecolor="black", alpha=0.4, align="center")
plt.xticks(tick_locations, df.x)
plt.show()
```
